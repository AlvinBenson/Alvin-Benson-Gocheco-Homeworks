```python
import re

# the function gets the numbers starting after 9 using regex
# the zfill adds a 0 to achieve the desired number of digits (including spaces)
# since a phone number format here has 11 digits with 2 spaces, it needed to be 13 "places" long
def phone_clean(x):
    number = re.search(r"[9]...........",x).group() 
    final_number = number.zfill(13)
    return final_number

# harder because there are many unique values
# using the help of regex I applied the function phone_clean which returns the final_number having the correct number
recweek['phonenumber'] = recweek['phonenumber'].apply(phone_clean)
recweek
```

```python
# Write code here

# People per Birthmonth sorted by calendar numeric month

# Grouping by birthmonth first and then counting .count()
# we need to make it to a dataframe so things will be easier to_frame()
# the birthmonth however gets transferred as the column name,so we change it to count then reset the index
birthmonth_gb = final_recweek.groupby("birthmonth")
birthmonth_df = birthmonth_gb["birthmonth"].count().to_frame().rename(columns={"birthmonth":"count"}).reset_index()
birthmonth_df


months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

# we want to organize the months in january to december order, with the help of the caregorical, I specify
# what is the order of the months I want to be, this makes the sort_values work because it has a basis
birthmonth_df['birthmonth'] = pd.Categorical(birthmonth_df['birthmonth'], categories=months, ordered=True)
birthmonth_df.sort_values(by='birthmonth',inplace=True)
birthmonth_df.reset_index(inplace=True,drop=True)
birthmonth_df
```

```python
# making a carrier df to have a smaller dataframe with the essentials
carrier_df = final_recweek[["gender","phonenumber"]].copy()
carrier_df
```

```python
import re
temporary = pd.DataFrame()

# passing the different phone numbers into this function
# if it starts with 0918 then its smart, if 0917 or 0927 then its globe
def carrier(x):
    if re.match(r"0918",x):
        return "Smart"
    elif re.match(r"0917",x):
        return "Globe"                
    elif re.match(r"0927",x):
        return "Globe"
    else:
        pass
    
temporary["carrier"] = carrier_df["phonenumber"].apply(carrier)
temporary[["gender"]] = carrier_df[["gender"]]
final_df = temporary.copy()
final_df
```
