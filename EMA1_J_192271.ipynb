```python
# Problem 1 Sample 1
inp = int(input("Enter number of lines: "))

print("<table>")
for i in range(inp):
    if i % 2 == 0:
        print('<tr class="lightyellowrow"><td>'+ str(i+1) + "</td><td>" + str((i+1)**2) + "</td></tr>")
    elif i % 2 == 1:
        print('<tr class="whiterow"><td>'+ str(i+1) + "</td><td>" + str((i+1)**2) + "</td></tr>")
    else:
        print("wrong")
print("</table>")
```

```python
#problem 3 Sample 1

def obedient(num):
    # num >= 6 because it needs to be a positive integer and the smallest possible is 1,2,3
    # mod 3 because because let the 3 numbers be n-1,n,n+1. Adding them up is 3n
    if num> 6 and num % 3 == 0:
        return True
    else:
        return False
    
print(obedient(123))
```

```python
#problem 4 Sample 1

def multtable(limit):
    # started at 1 because we don't want to start at 0
    # added 1 because of the nature of ranges to not include the upper limit
    for row in range(1,limit+1):
        for column in range(1,limit+1):
            # multiplying by the row, as we iterate in our range 1x1 1x2 1x3
            print(row*column,end="\t")
        print()

multtable(4)
```

```python
# Problem 2 Sample 3

Days_31 = [1,3,5,7,8,10,12]
Days_30 = [4,6,9,11]
leap_day = 0

day = int(input("Day: "))
month = int(input("Month: "))
year = int(input("Year: "))

# setting up the leap year
if year % 100 == 0 and year % 400==0:
    leap_day == 29
elif year % 4 ==0 and year % 100 != 0:
    leap_day == 29
else:
    leap_day == 28

# the months are not less than 1 or more than 12
if month < 1 or month >12:
    print("Date is not valid")
    
# checking the 31 and 30 days
elif month in Days_31 and (day>31 or day<1):
    print("Date is not valid")
elif month in Days_30 and (day>30 or day < 1):
    print("Date is not valid")

#checking if 28 or 29 days for february
elif month == 2 and (day>leap_day or day<1):
    print("Date is not valid")
else:

    
# if it passes the restrictions above, then we can print the actual date because it is acceptable 
    # for the day 1st,2nd,3rd
    if day == 1 or day == 21 or day == 31:
        day = str(day) + "st"
    elif day == 2 or day ==22:
        day = str(day) + "nd"
    elif day == 3 or day == 23:
        day = str(day) + "rd"
    else:
        day = str(day) + "th"
    
    # for the months
    if month == 1:
        month = "January"
    if month == 2:
        month = "February"
    if month == 3:
        month = "March"
    if month == 4:
        month = "April"
    if month == 5:
        month = "May"
    if month == 6:
        month = "June"
    if month == 7:
        month = "July"
    if month == 8:
        month = "August"
    if month == 9:
        month = "September"
    if month == 10:
        month = "October"
    if month == 11:
        month = "November"
    if month == 12:
        month = "December"

    print("It is the",day,"day of the month of",month,"of the year",year)
```

```python
# Problem 5 Sample 1
possible_answers = ["a","b","c"]
fire = 0
water = 0
earth = 0 

while True:
    Q1 = input("Do you prefer (a) the mountains, (b) the city, or (c) the beach? ")
    # insert into every Q1 so that it will go back, use continue (put everything in a while loop, or just the 5 questions)
    if Q1 not in possible_answers:
        print("Sorry, you inputted a wrong letter. Please start again.")
        continue

    Q2 = input("Do you prefer (a) yoghurt, (b) ice cream, or (c) gelato? ")
    if Q2 not in possible_answers:
        print("Sorry, you inputted a wrong letter. Please start again.")
        continue
        
    Q3 = input("Do you prefer (a) reading, (b) singing, or (c) writing? ")
    if Q3 not in possible_answers:
        print("Sorry, you inputted a wrong letter. Please start again.")
        continue
        
    Q4 = input("Do you prefer (a) sweet, (b) salty, or (c) spicy food? ")
    if Q4 not in possible_answers:
        print("Sorry, you inputted a wrong letter. Please start again.")
        continue
        
    Q5 = input("Do you prefer (a) cats, (b) dogs, or (c) fish? ")
    if Q5 not in possible_answers:
        print("Sorry, you inputted a wrong letter. Please start again.")
        continue
    else:
        # we break the loop once Q1 to Q5 have passed the restriction of not being in the possible answers
        break
        
#adding the elements for question 1 to 5
if Q1 == "a":
    fire += 1
elif Q1 == "b":
    earth += 1
elif Q1 == "c":
    water += 1
else:
    print("something went wrong in adding values to the elements")

if Q2 == "a":
    earth += 1
elif Q2 == "b":
    water += 1
elif Q2 == "c":
    fire += 1
else:
    print("something went wrong in adding values to the elements")
    
if Q3 == "a":
    water += 1
elif Q3 == "b":
    fire += 1
elif Q3 == "c":
    earth += 1
else:
    print("something went wrong in adding values to the elements")
    
if Q4 == "a":
    earth += 1
elif Q4 == "b":
    water += 1
elif Q4 == "c":
    fire += 1
else:
    print("something went wrong in adding values to the elements")

if Q5 == "a":
    earth += 1
elif Q5 == "b":
    fire += 1
elif Q5 == "c":
    water += 1
else:
    print("something went wrong in adding values to the elements")


# if there is a clear winner 
if water > fire and water > earth:
    x = "Water"
elif fire > water and fire > earth:
    x = "Fire"
elif earth > fire and earth > water:
    x = "Earth"
# if there are ties in order of (fire and earth), (fire and water), (earth and water)
elif fire == earth and (fire>water and earth>water):
    x = "Earth"
elif fire == water and (fire>earth and water>earth):
    x = "Fire"
elif earth == water and (earth>fire and water>fire):
    x = "Water"
else:
    print("Something went wrong with the comparisons")

"""
# in case we want to check the values
print("water",water)
print("fire",fire)
print("earth",earth)
"""

print("Your element is",x)
```
