```python
# problem 2
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

diagonaldown = set()
diagonalup = set()

def get_winner(board):
    # checking horizonatals
    for row in board:
        #print(row) we are printing rows
        if all([x=="X" for x in row]) == True:
            print("X")
        elif all([x=="O" for x in row]) == True:
            print("O")
        else:
            pass

    # checking verticals
    for column in zip(*board):
        # print(column) we are printing columns because of zip function
        if all([x=="X" for x in column]) == True:
            print("X")
        elif all([x=="O" for x in column]) == True:
            print("O")
        else:
            pass

    # diagonal down
    for i,v in enumerate(board):
        down = (board[i][i])
        diagonaldown.add(down)
    if len(diagonaldown) == 1 and "X" in diagonaldown :
        print("X")
    elif len(diagonaldown) == 1 and "O" in diagonaldown:
        print("O")
    else:
        pass

    # diagonal up
    for i,v in enumerate(board):
        down = (board[2-i][i])
        diagonalup.add(down)
    if len(diagonalup) == 1 and "X" in diagonalup :
        print("X")
    elif len(diagonalup) == 1 and "O" in diagonalup:
        print("O")
    else:
        pass
   
get_winner(board1)
print("board 1 above")
get_winner(board2)
print("board 2 above")
get_winner(board3)
print("board 3 above")
get_winner(board4)
print("board 4 above")
get_winner(board5)
print("board 5 above")
get_winner(board6)
print("board 6 above")

# There is nothing printed if no one wins, because that is the sample given for sample output when board6 is called
```

```python
#problem 3
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, 
    
# <-- Note that Python does not complain about this extra comma. This is perfectly okay.
       #     The creators of Python saw that dictionaries tend to be changed a lot, so allowing this is a
       #     convenience for developers who need to add elements when needed.
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

# entering the source and destination
source = input("Enter source: ")
destination = input("Enter destination: ")

def eta(source,destination):
    # list for the time stamps to be added, to be summed later
    time = list()
    
    # we want x to be false first, because we want to find where the source is as we iterate through the different keys
    # we only make it true, once we found it, so we can start looking for the destination part of the tuple
    x = False
    y = True
    
    # because a route can go from dlsu to admu, which does not follow the route upd-admu-dlsu, we can use a while loop that goes back
    # until we have found the destination part
    while y == True:
        # we want to look at the keys and values at the same time, so we can append easily
        for k,v in legs.items():
            if x == False:
                # if the user inputed a direct route, it will append that time (v) and break the loop immediately
                if k[0] == source and k[1]==destination:
                    time.append(v['travel_time_mins'])
                    y = False
                    break
                 
                # if the source part of tuple is found, but the destination part of the tuple is not the same then we make x true
                # so we can find k[1] for the destination tuple
                elif k[0] == source:
                    time.append(v['travel_time_mins']) 
                    x = True
                    continue
                else:
                    pass
            else:
                # we are going here if we already have found the source part of the touple, now finding where destination matches
                # we want to append the time as long as its not yet the destination, becuase we need to pass that route
                if k[1] != destination and x == True:
                    print(v['travel_time_mins'])
                    time.append(v['travel_time_mins'])
                # but once we found the destination it is time to break after appending the time
                elif x==True and k[1]==destination:
                    time.append(v['travel_time_mins'])
                    y = False
                    break
     
    # summing the list once the loop is broken, meaning source and destination have been found
    SUM = sum(time)
    print("ETA:",SUM)
    
eta(source,destination)

```
